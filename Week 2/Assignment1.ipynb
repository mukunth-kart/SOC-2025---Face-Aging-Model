1. Write a function to check if a number is a prime.


```python
def is_prime(n):
    # Hint: A prime number is greater than 1 and has no divisors other than 1 and itself.
    #Try looping from 2 to √n and check if any number divides n evenly.
    # Write your logic here
    if (n==1):
      print("1 is not a prime number")
      return
    for i in range(2,int(n**(1/2))):
      if (n%i==0):
        print(n, "is not a prime number.")
        return
    print(n, "is a prime number")
is_prime(13)
```

3. Write a function that takes a sentence as input, converts it to lowercase, and returns a dictionary with word counts.

```python
#Use str.lower().
#Use str.split() to break the sentence into words.
#Use a dictionary to count frequencies.
def count_words(sentence):
    # Write your logic here
    sentence = sentence.lower()
    sentence = sentence.split()
    count = dict()
    for i in sentence:
      if i in count:
        count[i] += 1
      else:
        count[i] = 1
    return count
sent = 'hi hi hello hello how are you'
count_words(sent)
```

4. Write a function to reverse a list manually.

```python
#Use a loop to pop elements from the end or build a new list in reverse.
def reverse_list(lst):
    # Write your logic here
    #return lst[::-1] easiest option but I don't think you want this
    l=list()
    length = len(lst)
    for i in range(length):
      l.append(lst.pop())
    return l
ex_lst = [1,2,3,45]
print(reverse_list(ex_lst))
```

1. Simulate a 2-layer neural network using NumPy

Input shape: (3,)

Hidden layer: 4 neurons (ReLU)

Output layer: 1 neuron (Sigmoid)

```python
import numpy as np

# Activation functions
def relu(x):
    return np.maximum(0, x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Step 1: Create input
x = np.array([0.5, -1.2, 2.0])

# Step 2: Initialize hidden layer weights and bias
W1 = np.random.randn(4, 3)
b1 = np.random.randn(4)

# Step 3: Compute hidden layer output
# TODO: Write the formula for z1 and a1
z1 = np.matmul(W1,x) + b1
a1 = relu(z1)

# Step 4: Output layer (1 neuron)
W2 = np.random.randn(1, 4)
b2 = np.random.randn(1)

# Step 5: Final output
# TODO: Write z2 and output using sigmoid
z2 = np.matmul(W2, a1) + b2
output = sigmoid(x)

print("Network output:", output)
```

Copy code


```python
class FeedForwardNN(nn.Module):
    def __init__(self):
        super().__init__()
        # TODO: Define Linear layers
        self.layer1 = nn.Linear(5, 10)
        self.layer2 = nn.Linear(10, 20)
        self.layer3 = nn.Linear(20, 30)

    def forward(self, x):
        x = torch.relu(self.layer1(x))
        x = torch.relu(self.layer2(x))
        x = self.layer3(x)
        return x

model = FeedForwardNN()
print(model)
```

Compute MSE Loss Between Two Tensors

```python
#get some random numbers

y_true = torch.randn(5)
y_pred = torch.randn(5)


#write function here
def MSE(y_true, y_pred, n):
  loss = np.matmul(y_true, y_true) + np.matmul(y_pred, y_pred) - 2*np.matmul(y_true,y_pred)
  loss = loss/n
  loss = (loss)**(1/2)
  return loss
loss = MSE(y_true, y_pred, 5)
print("MSE Loss:", loss.item())
```
