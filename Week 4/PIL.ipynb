#**Image Module**

```python
from PIL import Image #import image module from PIL library
import matplotlib.pyplot as plt
```

```python
img = Image.open('x.png')
print(img)
```

```python
plt.imshow(img) #Does not work in colab only works on local libss
plt.title("original")
plt.show()
```

```python
plt.imshow(img) #Does not work in colab only works on local libss
plt.title("original")
plt.axis('off')
plt.show()
```

```python
img_new = Image.new("RGB", (100,100),(255, 200,0) ) #(mode, size, colour)
plt.imshow(img_new)                                 #RGB has three color channels
plt.axis("off")
plt.show()
```

```python
img_new = Image.new("RGB", (100,100),'yellow' ) #(mode, size, colour) #Limited choices though
plt.imshow(img_new)
plt.axis("off")
plt.show()
```

Rotating an image

```python
img_r = img.rotate(angle =60, expand=True, fillcolor='Black', resample=Image.BICUBIC, center=(100,100)) #(angle(degrees), expand, fillcolor, resample, center)
plt.imshow(img_r)                  #Resample is for high quality
plt.axis("off")                    #center is for centre of rotation
plt.title("rotated")
plt.show()
```

Resizing an image

```python
resized = img.resize((img.width,img.height), resample = Image.LANCZOS, box=(20,20,100,100))
plt.imshow(resized) #Box chooses a specific portions
plt.title("resized")
plt.axis("off")
plt.show()
```

Blend two Images

```python
blended = Image.blend(img, resized, alpha=0.5)
plt.imshow(blended)
plt.title("blended")
plt.axis("off")
plt.show()
```

Cropping an image

```python
img_crp = img.crop(box=(20,20,200,175))
plt.imshow(img_crp)
plt.title("cropped")
plt.axis("off")
plt.show()
```
